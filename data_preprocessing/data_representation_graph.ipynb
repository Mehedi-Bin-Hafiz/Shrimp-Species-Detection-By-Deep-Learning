```python
import tensorflow.keras as keras
from keras.preprocessing.image import load_img
import numpy as np
from keras.preprocessing import image
from matplotlib import  pyplot as plt
import time
import os
import cv2
from skimage import io
import time
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams.update({'font.size': 12})
```

```python
save_here = 'D:\\Research database\\Shrimp Recognition\\training'
main_classes = os.listdir(save_here) # list of directory
print(main_classes)
len_main_classes = len(main_classes)
print('total class ',len_main_classes)
```

```python
image_classes = os.listdir(save_here)
counter_lis = list()
for class_name, category in enumerate(image_classes):
    save_dir = os.path.join(save_here,category)
    category_dir =  os.path.join(save_here,category)
    counter_lis.append(len(os.listdir(category_dir)))
```

```python
figure(figsize=(6, 4), dpi=300)
explode = (0.013,0.013,0.013,0.013)
plt.pie(counter_lis, explode=explode, labels=main_classes, autopct='%1.1f%%',)
plt.savefig('trainingDatasetRepresentation.jpeg') # need to call before calling show
plt.show()
```
