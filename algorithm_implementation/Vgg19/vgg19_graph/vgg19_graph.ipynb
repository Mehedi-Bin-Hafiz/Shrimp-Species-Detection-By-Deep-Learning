```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
# plt.rcParams["figure.figsize"] = [12, 6]
plt.rcParams.update({'font.size': 14})
plt.rcParams["font.family"] = "Times New Roman"
figure(figsize=(6, 4), dpi=500)

############################ Model 1 ####################3
model_1 = pd.read_excel('../vgg19_history.xlsx',engine='openpyxl')

model_1TAccu = model_1['accuracy'].values.tolist()
model_1VAccu = model_1['val_accuracy'].values.tolist()
model_1TLoss = model_1['loss'].values.tolist()
model_1VLoss =model_1['val_loss'].values.tolist()


axes = plt.axes()
plt.plot(range(1,len(model_1TAccu)+1),model_1TAccu,color='green',linewidth=1.5)
plt.plot(range(1,len(model_1TAccu)+1),model_1VAccu,color='red',linewidth=1.5)
# axes.set_yticks([0.8,0.85,0.9,0.95,1.00,1.050,1.100])
plt.grid()
axes.set_yticks([0.3,0.4,0.5,0.6,0.7,0.8,0.9,1,1.1,1.2,1.3])
axes.set_xticks([x for x in range(0,35) if x%5 == 0])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
# plt.ylabel('Loss')
plt.legend(['Training Accuracy', 'Validation Accuracy'])
plt.savefig('vgg19  tain vs validation accuracy.jpeg')
plt.show()
```

```python
figure(figsize=(6, 4), dpi=500)
axes = plt.axes()
plt.plot(range(1,len(model_1TLoss)+1),model_1TLoss,color='green',linewidth=1.5)
plt.plot(range(1,len(model_1TLoss)+1),model_1VLoss,color='red',linewidth=1.5)
plt.grid()
axes.set_yticks([x for x in range(0,12) if x%2   == 0])
axes.set_xticks([x for x in range(0,35) if x%5 == 0])
plt.xlabel('Epochs')
plt.ylabel('Loss')
# plt.ylabel('Loss')
plt.legend(['Training Loss', 'Validation Loss'])
plt.savefig('vgg19 tain vs validation loss.jpeg')
plt.show()
```
