```python
import pandas as pd
import tensorflow.keras as keras
from keras.utils import load_img
import numpy as np
from keras.preprocessing import image
from matplotlib import  pyplot as plt
import time
import os
import cv2
import time
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras 
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
model = keras.models.load_model('../vgg19_model.h5') 
```

```python
testing_dataset = 'D:\\Research Dataset\\Shrimp Recognition\\testing\\' 
main_classes = os.listdir(testing_dataset) 
print(main_classes)
len_main_classes = len(main_classes)
print('total class ',len_main_classes)
```

```python
real_class_list = []
prediction_class_list = []
image_classes = os.listdir(testing_dataset)
count = 0

for real_class, category in enumerate(image_classes):
    category_dir = os.path.join(testing_dataset, category)
    start_time = time.time()
    for img_name in os.listdir(category_dir):
        img_path = os.path.join(category_dir, img_name)
        single_image = load_img(img_path, target_size=(224, 224), color_mode='rgb')
        img_array = image.img_to_array(single_image) / 255.0
        img_batch = np.expand_dims(img_array, axis=0)
        prediction = np.argmax(model.predict(img_batch), axis=1)[0]
        real_class_list.append(real_class)
        prediction_class_list.append(prediction)

        count += 1
        if count % 100 == 0:
            print(f"{count} images prediction completed. Time taken: {time.time() - start_time:.2f} s")
            start_time = time.time()
```

```python
df_dict = {'real':real_class_list,'prediction':prediction_class_list}
df = pd.DataFrame(df_dict)
df.to_excel('model_2_realVs_pred.xlsx',index = False)
```

```python
from sklearn.metrics import classification_report
target_names = ["Bagda","Deshi","Golda",'Horina']
print(classification_report(real_class_list, prediction_class_list, target_names=target_names))
```
