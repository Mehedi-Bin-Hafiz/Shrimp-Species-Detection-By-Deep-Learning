```python
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras 
model = keras.models.load_model('../Shrimp_model_2.h5')
```

```python
np.set_printoptions(suppress=True) #golda0.99 and Horina=0.99 detected accurately
classes = ["Bagda","Deshi","Golda",'Horina']
img_path = "golda_test.png"      
img_array = None
img = image.load_img(img_path, target_size=(224,224))
img_array = image.img_to_array(img)/255
plt.imshow(img_array)
img_batch = np.expand_dims(img_array, axis=0)
prediction = model.predict_proba(img_batch)
print(prediction)
max_ind = np.argmax(prediction)
print("\n",classes[max_ind])
```
